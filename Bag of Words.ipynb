```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("spam.csv")
df.head()
```

```python
df.Category.value_counts()
```

```python
df['spam'] = df['Category'].apply(lambda x:1 if x=='spam' else 0)
```

```python
df.shape
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(df.Message,df.spam,test_size =0.2) #x-messsage y-spam
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train[:4]
```

```python
y_train[:4]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
x_train_cv = v.fit_transform(x_train.values)
x_train_cv
```

```python
x_train_cv .shape
```

```python
v.get_feature_names_out() #how to find this method
```

```python
dir(v) #all the methods
```

```python
v.vocabulary_
```

```python
x_train_np = x_train_cv.toarray()
x_train_np[:4]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_cv,y_train)
```

```python
x_test_cv = v.transform(x_test)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(x_test_cv)
print(classification_report(y_test,y_pred))
```

```python
emails = [
    'Hey Hari, can we get together to watch football game tommorrow?',
    'Upto 20% discount on parking,exclusive offer just for you. Dont miss this reward!']
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
print(classification_report(y_test,y_pred))
```
